```python
import numpy as np
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import os
import sys
import re
import math
import holoviews as hv
import panel as pn
import bisect
hv.extension('bokeh', 'matplotlib')
from IPython.display import display
from ipyfilechooser import FileChooser
import warnings
from scipy.stats import zscore
import matplotlib.cm as cm
import matplotlib.colors as mcolors
from scipy.signal import resample
from scipy.stats import sem
import json

%matplotlib widget
warnings.filterwarnings("ignore")
#%reset

minian_path = "C:/Users/manip/Documents/MB/minian/"
def remove_outliers_avg_filter(data):
    data = np.array(data, dtype=float)  # Ensure NumPy array with float type
    filtered_data = np.copy(data)  # Copy to avoid modifying original data
    for i in range(len(data)):
        if not np.isnan(data[i]):  # Skip valid values
            continue
        # Find the closest previous non-NaN value
        prev_idx = i - 1
        while prev_idx >= 0 and np.isnan(data[prev_idx]):
            prev_idx -= 1        
        # Find the closest next non-NaN value
        next_idx = i + 1
        while next_idx < len(data) and np.isnan(data[next_idx]):
            next_idx += 1
        # Compute average if both values exist
        if prev_idx >= 0 and next_idx < len(data):
            filtered_data[i] = (data[prev_idx] + data[next_idx]) / 2
        # If neither exists, NaN remains
    return filtered_data

def find_closest_index_sorted(arr, target):
    idx = bisect.bisect_left(arr, target)  # Find the insertion point
    if idx == 0:
        return 0
    if idx == len(arr):
        return len(arr) - 1
    before = idx - 1
    after = idx
    return before if abs(arr[before] - target) <= abs(arr[after] - target) else after

def detect_longest_lowest_sequence(arr, margin=0):
    min_val = np.nanmin(arr)  # Find minimum value
    threshold = min_val + margin  # Define threshold based on margin    
    # Get indices where values are within the threshold
    min_indices = np.where(arr <= threshold)[0]
    # Identify consecutive sequences
    longest_sequence = None
    if len(min_indices) > 0:
        start = min_indices[0]
        max_duration = 0  # Track longest duration        
        for i in range(1, len(min_indices)):
            if min_indices[i] != min_indices[i - 1] + 1:  # Not consecutive
                duration = min_indices[i - 1] - start + 1
                if duration > max_duration:
                    max_duration = duration
                    longest_sequence = (start, min_indices[i - 1], duration)
                start = min_indices[i]  # Reset start index        
        # Check last detected sequence
        duration = min_indices[-1] - start + 1
        if duration > max_duration:
            longest_sequence = (start, min_indices[-1], duration)
    return min_val, threshold, longest_sequence
```

```python
from pathlib import Path
import os
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from mpl_toolkits.axes_grid1 import make_axes_locatable

# Dossier de base
mouse_date = "2025_06_05"
hour = "11_19_07"
mouse_id = "YL"
dpath = Path(fr"S:/forgetting/Clementine/CheeseboardExperiment/Juin/YL/Cheeseboard/2025_06_05/11_19_07/My_V4_Miniscope")

# Trouver le bon dossier DLC
dlcpath = dpath.parent / "My_First_WebCam"
if not dlcpath.exists():
    variants = [
        'My_First_WebCam', 'My_first_WebCam', 'My_First_Webcam', 'My_first_webcam',
        'MyFirstWebCam', 'myfirstwebcam', 'my_first_webcam', 'webcam', 'WebCam'
    ]
    for variant in variants:
        test_path = dpath.parent / variant
        if test_path.exists():
            dlcpath = test_path
            break

# Charger le fichier CSV (au lieu de HDF5)
csv_files = [f for f in os.listdir(dlcpath) if f.endswith('.csv') and 'timeStamps' not in f]
csv_file = csv_files[0]
csv_path = dlcpath / csv_file
df = pd.read_csv(csv_path, header=[1, 2], index_col=0)

# Charger les timestamps s'ils existent
timestamps_path = dlcpath / "timeStamps.csv"
if timestamps_path.exists():
    timestamps = pd.read_csv(timestamps_path)
    frame_rate = round(1 / (np.mean(np.diff(timestamps.iloc[:, 1])) / 1000))
else:
    frame_rate = 30  # par défaut

# Extraction de la position (prend la première étiquette disponible)
first_label = df.columns.get_level_values(0)[0]
X = df[first_label]["x"].copy()
Y = df[first_label]["y"].copy()
likelihood = df[first_label]["likelihood"]

# Masquer les points de faible confiance
X[likelihood < 0.9] = np.nan
Y[likelihood < 0.9] = np.nan
X = X.to_numpy()
Y = Y.to_numpy()

# Coordonnées et conversion
pixel_to_cm = 2.25
table_center_x, table_center_y = 313, 283
table_radius = 270 / 2

def calculate_relative_distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

def calculate_distance_run(x_coords, y_coords):
    distances = np.sqrt(np.diff(x_coords)**2 + np.diff(y_coords)**2)
    distances = np.nan_to_num(distances, nan=0.0)
    return np.nansum(distances) / pixel_to_cm, distances

def find_long_non_nan_sequences(arr, min_length=5):
    mask = ~np.isnan(arr)
    diff = np.diff(np.concatenate(([0], mask.astype(int), [0])))
    starts = np.where(diff == 1)[0]
    ends = np.where(diff == -1)[0]
    sequences = [arr[start:end] for start, end in zip(starts, ends) if (end - start) > min_length]
    return sequences

def remove_outliers_median_filter(data, window=4):
    data = np.array(data, dtype=float)
    filtered_data = np.copy(data)
    half_window = window // 2
    for i in range(len(data)):
        start = max(0, i - half_window)
        end = min(len(data), i + half_window + 1)
        local_values = data[start:end]
        median_value = np.nanmedian(local_values) if not np.all(np.isnan(local_values)) else np.nan
        if not np.isnan(data[i]):
            filtered_data[i] = median_value
    return filtered_data

# Enlever les points hors plateau
for i, x in enumerate(X):
    y = Y[i]
    if calculate_relative_distance(x, y, table_center_x, table_center_y) >= table_radius:
        X[i] = np.nan
        Y[i] = np.nan

# Détection du début
x_start = find_long_non_nan_sequences(X)[0][0]
y_start = find_long_non_nan_sequences(Y)[0][0]
start_frame = np.where(X == x_start)[0][0]
X[:start_frame] = np.nan
Y[:start_frame] = np.nan

X = remove_outliers_median_filter(X)
Y = remove_outliers_median_filter(Y)

# Dernier frame valide
for i in range(len(X) - 1, 0, -1):
    if not np.isnan(X[i]) and not np.isnan(X[i - 1]):
        last_frame = i
        break

# Durée du trial
if timestamps_path.exists():
    start_time = timestamps.iloc[start_frame, 1] / 1000
    end_time = timestamps.iloc[last_frame, 1] / 1000
    duration_trial = end_time - start_time
else:
    duration_trial = (last_frame - start_frame) / frame_rate

print(f"Total trial duration: {round(duration_trial)} sec")

# Distance totale
total_distance, speeds = calculate_distance_run(X[start_frame:last_frame], Y[start_frame:last_frame])
print(f"Total distance run: {round(total_distance)} cm")

# Vitesse moyenne
avg_speed = total_distance / duration_trial
print(f"Average speed: {round(avg_speed, 2)} cm/s")

# Plot amélioré avec vitesse + start + end
individual_x_nonan = X[start_frame:last_frame]
individual_y_nonan = Y[start_frame:last_frame]
speeds = np.append(speeds, speeds[-1])  # pour aligner la taille avec les coordonnées

points = np.array([individual_x_nonan, individual_y_nonan]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)

fig, ax = plt.subplots(figsize=(5, 5))

# Segments colorés par vitesse
lc = LineCollection(segments, cmap='viridis', norm=plt.Normalize(0, np.nanmax(speeds)))
lc.set_array(speeds)
lc.set_linewidth(2)
line = ax.add_collection(lc)

# Start & End
start_scatter = ax.scatter(individual_x_nonan[0], individual_y_nonan[0], color='green', s=100, label='Start')
end_scatter = ax.scatter(individual_x_nonan[-1], individual_y_nonan[-1], color='red', s=100, label='End')

# Cercle plateau
ax.add_patch(plt.Circle((table_center_x, table_center_y), table_radius, color='black', fill=False))

# Mise en forme
ax.set_aspect('equal')
ax.invert_yaxis()
ax.set_title('Mouse Path with Speed Coloring')
ax.set_xlabel('X Position')
ax.set_ylabel('Y Position')

# Axes supplémentaires pour colorbar et légende externe
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
legend_ax = divider.append_axes("right", size="15%", pad=0.3)
cbar = plt.colorbar(lc, cax=cax)
cbar.set_label('Speed (px/frame)')

# Légende Start/End à droite
legend_ax.axis('off')
legend_ax.legend(
    handles=[start_scatter, end_scatter],
    labels=['Start', 'End'],
    loc='center left',
    frameon=False
)

plt.tight_layout()

# Sauvegarde automatique
output_path = dlcpath / f"{mouse_id}_{mouse_date}_{hour}_path_plot.png"
plt.savefig(output_path, dpi=300)
print(f"Figure saved to: {output_path}")

plt.show()

```

```python
import numpy as np
import pandas as pd
import xarray as xr
from pathlib import Path

def remove_outliers_avg_filter(data, window_size=5, z_thresh=3):
    """
    Nettoie les données en supprimant les valeurs aberrantes par un filtre moyen local.
    """
    data = np.array(data)
    filtered = data.copy()
    kernel = np.ones(window_size) / window_size
    moving_avg = np.convolve(filtered, kernel, mode='same')
    
    def moving_std(x, w):
        cumsum = np.cumsum(np.insert(x**2, 0, 0)) 
        cumsum_mean = np.cumsum(np.insert(x, 0, 0))
        mean_sq = (cumsum[w:] - cumsum[:-w]) / w
        mean = (cumsum_mean[w:] - cumsum_mean[:-w]) / w
        return np.sqrt(mean_sq - mean**2)
    
    half_w = window_size // 2
    std_local = np.full_like(data, np.nan)
    std_local[half_w:-half_w] = moving_std(data, window_size)
    std_local[:half_w] = np.std(data[:window_size])
    std_local[-half_w:] = np.std(data[-window_size:])
    
    z_scores = np.abs(data - moving_avg) / std_local
    outliers = z_scores > z_thresh
    filtered[outliers] = moving_avg[outliers]
    return filtered

def find_closest_index_sorted(array, value):
    array = np.asarray(array)
    idx = np.searchsorted(array, value, side="left")
    if idx > 0 and (idx == len(array) or np.abs(value - array[idx-1]) < np.abs(value - array[idx])):
        return idx - 1
    else:
        return idx

def calculate_distance_run(x, y):
    """
    Calcule la distance cumulée et la vitesse instantanée à partir de coordonnées x, y.
    """
    dx = np.diff(x)
    dy = np.diff(y)
    dist = np.sqrt(dx**2 + dy**2)
    total_distance = np.nansum(dist)
    speed = np.concatenate(([0], dist))  # vitesse instantanée (1 frame delta)
    return total_distance, speed

# --- Simulation des timestamps (en ms) ---
session_duration_s = 303
sampling_rate_hz = 1000  # 1000 Hz pour miniscope
tsmini = np.linspace(0, session_duration_s * 1000, session_duration_s * sampling_rate_hz, dtype=float)
minian_freq = round(1 / np.mean(np.diff(tsmini / 1000)))
print('🎞️  Miniscope sample rate =', minian_freq, 'Hz')

# Simulations variables pour exemple
start_time = 0.0
end_time = session_duration_s
start_frame = 0
last_frame = 1000
frame_rate = 30  # webcam rate (exemple)

# Simuler positions avec bruit et quelques outliers
np.random.seed(42)
individual_x = np.cumsum(np.random.randn(last_frame)) + 100
individual_y = np.cumsum(np.random.randn(last_frame)) + 50
# injecter un outlier
individual_x[500] += 50
individual_y[500] -= 40

# Simuler traces calcium (10 unités, 1000 frames)
C = xr.DataArray(np.random.rand(10, last_frame), dims=("unit", "frame"))

# Nettoyer positions
individual_x_nonan = remove_outliers_avg_filter(individual_x)
individual_y_nonan = remove_outliers_avg_filter(individual_y)

# Downsample calcium pour coller à la longueur des positions
new_len = len(individual_x_nonan)
old_indices = np.linspace(0, 1, C.shape[1])
new_indices = np.linspace(0, 1, new_len)
Cds_all = np.array([np.interp(new_indices, old_indices, C.values[i, :]) for i in range(C.shape[0])])

# Calcul distance et vitesse
total_distance, speed = calculate_distance_run(
    individual_x_nonan[start_frame:last_frame],
    individual_y_nonan[start_frame:last_frame]
)

print(f"Distance totale parcourue (approx): {total_distance:.2f} unités de position")
print(f"Vitesse (extrait des 10 premières frames): {speed[:10]}")

# --- Suppression de toute référence à reward ---

# Simulation du path timestamps
timestamps_path = Path("timestamps.csv")
timestamps = pd.DataFrame({
    "frame": np.arange(last_frame),
    "timestamp_ms": tsmini[:last_frame]
})

# Simuler indices de début et fin d'un événement (exemple)
start_eating_frame = 100
end_eating_frame = 200

if timestamps_path.exists():
    start_eating_time = timestamps.iloc[start_eating_frame, 1] / 1000
    end_eating_time = timestamps.iloc[end_eating_frame, 1] / 1000
    duration_eating = end_eating_time - start_eating_time
else:
    start_eating_time = start_eating_frame / frame_rate
    end_eating_time = end_eating_frame / frame_rate
    duration_eating = (end_eating_frame - start_eating_frame) / frame_rate

print(f"Début de l'événement : {start_eating_time:.2f}s")
print(f"Fin de l'événement : {end_eating_time:.2f}s")
print(f"Durée de l'événement : {duration_eating:.2f}s")


```

```python
#DONNEES NON TRIEES, POUR UNE SEULE SESSION

import os
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
from matplotlib.collections import LineCollection
import xarray as xr
import math

# ======================== Paramètres ========================
nr = 5  # Numéro du neurone à visualiser (commence à 1)
dpath = Path("S:/forgetting/Clementine/CheeseboardExperiment/Juin/YL/Cheeseboard/2025_06_05/11_19_07/My_V4_Miniscope/")
minianversion = "minian"

# ======================== Définition des chemins ========================
folderMouse = dpath / minianversion
StampsMiniscopeFile = dpath / "timeStamps.csv"

A_zarr_path = folderMouse / "A.zarr"
C_zarr_path = folderMouse / "C.zarr"

# ======================== Vérification de l'existence des fichiers ========================
print("🔍 Vérification des fichiers...")
if not dpath.exists():
    raise FileNotFoundError(f"Le dossier principal n'existe pas : {dpath}")
if not folderMouse.exists():
    raise FileNotFoundError(f"Le dossier minian n'existe pas : {folderMouse}")
if not A_zarr_path.exists():
    raise FileNotFoundError(f"Le fichier A.zarr n'existe pas : {A_zarr_path}")
if not C_zarr_path.exists():
    raise FileNotFoundError(f"Le fichier C.zarr n'existe pas : {C_zarr_path}")
if not StampsMiniscopeFile.exists():
    raise FileNotFoundError(f"Le fichier timeStamps.csv n'existe pas : {StampsMiniscopeFile}")
print("✅ Tous les fichiers requis existent")

# ======================== Chargement des données calcium ========================
print("📊 Chargement des données calcium...")
try:
    A_ds = xr.open_zarr(A_zarr_path)
    C_ds = xr.open_zarr(C_zarr_path)
    print("✅ Fichiers Zarr chargés avec succès")
except Exception as e:
    print(f"❌ Erreur lors du chargement des fichiers Zarr : {e}")
    print(f"Contenu du dossier {folderMouse}:")
    for item in folderMouse.iterdir():
        print(f"  - {item}")
    raise

try:
    tsmini = pd.read_csv(StampsMiniscopeFile)['Time Stamp (ms)']
    minian_freq = round(1 / np.mean(np.diff(np.array(tsmini) / 1000)))
    print('🎞️  Miniscope sample rate =', minian_freq, 'Hz')
except Exception as e:
    print(f"❌ Erreur lors du chargement des timestamps : {e}")
    raise

try:
    A = A_ds['A']
    C = C_ds['C']
    print("✅ Matrices A et C extraites sans filtre")
except KeyError as e:
    print(f"❌ Clé manquante dans les datasets Zarr : {e}")
    print("Clés disponibles dans A_ds:", list(A_ds.keys()))
    print("Clés disponibles dans C_ds:", list(C_ds.keys()))
    raise

Cds_all = C.values
n_units = C.shape[0]
print(f"➡️ Nombre total d'unités : {n_units}")
if nr < 1 or nr > n_units:
    raise ValueError(f"nr doit être entre 1 et {n_units}")

# ======================== Données de position (DeepLabCut) ========================
print("🎯 Chargement des données de position...")
dlcpath = dpath.parent / "My_First_WebCam"
if not dlcpath.exists():
    print(f"⚠️ Dossier {dlcpath} non trouvé, recherche de variantes...")
    variants = [
        'My_First_WebCam', 'My_first_WebCam', 'My_First_Webcam', 'My_first_webcam',
        'MyFirstWebCam', 'myfirstwebcam', 'my_first_webcam', 'webcam', 'WebCam']
    for variant in variants:
        test_path = dpath.parent / variant
        if test_path.exists():
            dlcpath = test_path
            print(f"✅ Trouvé : {dlcpath}")
            break
    else:
        raise FileNotFoundError(f"Aucun dossier de données DLC trouvé dans {dpath.parent}")

try:
    csv_files = [f for f in os.listdir(dlcpath) if f.endswith('.csv') and 'timeStamps' not in f]
    if not csv_files:
        raise FileNotFoundError(f"Aucun fichier CSV trouvé dans {dlcpath}")
    csv_path = dlcpath / csv_files[0]
    print(f"📁 Utilisation du fichier : {csv_files[0]}")
    df = pd.read_csv(csv_path, header=[1, 2], index_col=0)
    print("✅ Données DLC chargées")
except Exception as e:
    print(f"❌ Erreur lors du chargement des données DLC : {e}")
    raise

first_label = df.columns.get_level_values(0)[0]
X = df[first_label]["x"].copy()
Y = df[first_label]["y"].copy()
likelihood = df[first_label]["likelihood"]

X[likelihood < 0.9] = np.nan
Y[likelihood < 0.9] = np.nan
X = X.to_numpy()
Y = Y.to_numpy()
print(f"📍 Coordonnées extraites : {len(X)} points")

def calculate_relative_distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

table_center_x, table_center_y = 313, 283
table_radius = 270 / 2
initial_valid_points = np.sum(~np.isnan(X))
for i, x in enumerate(X):
    y = Y[i]
    if not np.isnan(x) and not np.isnan(y):
        if calculate_relative_distance(x, y, table_center_x, table_center_y) >= table_radius:
            X[i] = np.nan
            Y[i] = np.nan
final_valid_points = np.sum(~np.isnan(X))
print(f"🎯 Points supprimés hors cercle : {initial_valid_points - final_valid_points}")

def find_long_non_nan_sequences(arr, min_length=5):
    mask = ~np.isnan(arr)
    diff = np.diff(np.concatenate(([0], mask.astype(int), [0])))
    starts = np.where(diff == 1)[0]
    ends = np.where(diff == -1)[0]
    return [arr[start:end] for start, end in zip(starts, ends) if (end - start) > min_length]

def remove_outliers_median_filter(data, window=4):
    data = np.array(data, dtype=float)
    filtered_data = np.copy(data)
    half_window = window // 2
    for i in range(len(data)):
        start = max(0, i - half_window)
        end = min(len(data), i + half_window + 1)
        local_values = data[start:end]
        median_value = np.nanmedian(local_values) if not np.all(np.isnan(local_values)) else np.nan
        if not np.isnan(data[i]):
            filtered_data[i] = median_value
    return filtered_data

x_sequences = find_long_non_nan_sequences(X)
y_sequences = find_long_non_nan_sequences(Y)
if not x_sequences or not y_sequences:
    raise ValueError("Pas assez de données valides pour analyser la trajectoire")
x_start = x_sequences[0][0]
y_start = y_sequences[0][0]
start_frame = np.where(X == x_start)[0][0]
X[:start_frame] = np.nan
Y[:start_frame] = np.nan
print(f"🚀 Début d'analyse à la frame : {start_frame}")

X = remove_outliers_median_filter(X)
Y = remove_outliers_median_filter(Y)
last_frame = len(X) - 1
for i in range(len(X) - 1, 0, -1):
    if not np.isnan(X[i]) and not np.isnan(X[i - 1]):
        last_frame = i
        break
print(f"🏁 Fin d'analyse à la frame : {last_frame}")
if last_frame <= start_frame:
    raise ValueError("Pas assez de données valides pour l'analyse")

individual_x_nonan = X[start_frame:last_frame]
individual_y_nonan = Y[start_frame:last_frame]
if start_frame >= Cds_all.shape[1] or last_frame > Cds_all.shape[1]:
    print(f"⚠️ Ajustement des indices: données calcium={Cds_all.shape[1]}, position={len(X)}")
    last_frame = min(last_frame, Cds_all.shape[1])
    individual_x_nonan = X[start_frame:last_frame]
    individual_y_nonan = Y[start_frame:last_frame]

Cds_nr = Cds_all[nr - 1, start_frame:last_frame]
print(f"📊 Données finales : {len(individual_x_nonan)} points pour le neurone {nr}")

# ======================== Visualisation ========================
print("🎨 Création de la visualisation...")
fig, ax = plt.subplots(figsize=(8, 8))
if len(Cds_nr) > 0 and Cds_nr.min() != Cds_nr.max():
    norm = mcolors.Normalize(vmin=Cds_nr.min(), vmax=Cds_nr.max())
else:
    norm = mcolors.Normalize(vmin=0, vmax=1)
cmap = cm.jet

for i in range(len(individual_x_nonan) - 1):
    if not np.isnan(individual_x_nonan[i]) and not np.isnan(individual_y_nonan[i]) and \
       not np.isnan(individual_x_nonan[i + 1]) and not np.isnan(individual_y_nonan[i + 1]):
        color = cmap(norm(Cds_nr[i]))
        ax.plot([individual_x_nonan[i], individual_x_nonan[i + 1]],
                [individual_y_nonan[i], individual_y_nonan[i + 1]],
                color=color, linewidth=1.0)

circle = plt.Circle((table_center_x, table_center_y), table_radius, 
                   color='black', fill=False, linewidth=2, linestyle='--')
ax.add_patch(circle)

sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, fraction=0.046, pad=0.04)
cbar.set_label('Activité calcique', fontsize=12)

ax.set_title(f"Trajectoire du Neuron #{nr}", fontsize=14, fontweight='bold')
ax.set_xlabel('Position X (pixels)', fontsize=12)
ax.set_ylabel('Position Y (pixels)', fontsize=12)
ax.set_aspect('equal')
ax.invert_yaxis()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
print("✅ Analyse terminée avec succès !")
```

```python
#STATS
import os
from pathlib import Path
import pandas as pd
import numpy as np
import xarray as xr
import json
from scipy.stats import ks_2samp, fisher_exact
from IPython.display import display

# ======== PARAMÈTRES ========
dpath = Path(r"S:\forgetting\Clementine\CheeseboardExperiment\Juin\YL\Cheeseboard\2025_06_05\11_19_07\My_V4_Miniscope")
minianversion = "minian"

folderMouse = dpath / minianversion
A_zarr_path = folderMouse / "A.zarr"
C_zarr_path = folderMouse / "C.zarr"
todrop_path = dpath / "TodropFileAB.json"
head_orientation_file = dpath / "headOrientation.csv"

# ======== Vérification des fichiers ========
for path in [dpath, folderMouse, A_zarr_path, C_zarr_path, todrop_path, head_orientation_file]:
    if not path.exists():
        raise FileNotFoundError(f"Fichier manquant : {path}")

print("✅ Tous les fichiers requis existent")

# ======== Chargement données calcium ========
A_ds = xr.open_zarr(A_zarr_path)
C_ds = xr.open_zarr(C_zarr_path)

with open(todrop_path, 'r') as f:
    to_drop = json.load(f)

all_units = np.arange(C_ds['C'].shape[0])
kept_units = np.setdiff1d(all_units, to_drop)
C_filtered = C_ds['C'].values[kept_units]

# ======== Chargement et conversion orientation tête (quaternion → angles) ========
head_df = pd.read_csv(head_orientation_file)

qw = head_df['qw'].to_numpy()
qx = head_df['qx'].to_numpy()
qy = head_df['qy'].to_numpy()
qz = head_df['qz'].to_numpy()

def quaternion_to_euler(qw, qx, qy, qz):
    sinr_cosp = 2 * (qw * qx + qy * qz)
    cosr_cosp = 1 - 2 * (qx * qx + qy * qy)
    roll = np.arctan2(sinr_cosp, cosr_cosp)

    sinp = 2 * (qw * qy - qz * qx)
    sinp = np.clip(sinp, -1.0, 1.0)
    pitch = np.arcsin(sinp)

    siny_cosp = 2 * (qw * qz + qx * qy)
    cosy_cosp = 1 - 2 * (qy * qy + qz * qz)
    yaw = np.arctan2(siny_cosp, cosy_cosp)

    return roll, pitch, yaw

roll, pitch, yaw = quaternion_to_euler(qw, qx, qy, qz)
roll_deg = np.degrees(roll)
pitch_deg = np.degrees(pitch)
yaw_deg = np.degrees(yaw)

def remove_outliers_iqr(data, factor=1.5):
    q1 = np.nanpercentile(data, 25)
    q3 = np.nanpercentile(data, 75)
    iqr = q3 - q1
    lower_bound = q1 - factor * iqr
    upper_bound = q3 + factor * iqr
    filtered = data.copy()
    filtered[(data < lower_bound) | (data > upper_bound)] = np.nan
    return filtered

roll_deg = remove_outliers_iqr(roll_deg)
pitch_deg = remove_outliers_iqr(pitch_deg)
yaw_deg = remove_outliers_iqr(yaw_deg)

valid_mask = ~(np.isnan(roll_deg) | np.isnan(pitch_deg) | np.isnan(yaw_deg))
valid_indices = np.where(valid_mask)[0]

if len(valid_indices) == 0:
    raise ValueError("Aucune donnée d'orientation valide")

start_frame = valid_indices[0]
last_frame = valid_indices[-1] + 1

roll_deg = roll_deg[start_frame:last_frame]
pitch_deg = pitch_deg[start_frame:last_frame]
yaw_deg = yaw_deg[start_frame:last_frame]

if last_frame > C_filtered.shape[1]:
    last_frame = C_filtered.shape[1]

C_filtered = C_filtered[:, start_frame:last_frame]

# ======== Fonctions de test de modulation activité vs angle tête ========

def compute_spatial_info(activity, angle, nbins=10):
    bins = np.linspace(np.nanmin(angle), np.nanmax(angle), nbins + 1)
    inds = np.digitize(angle, bins) - 1
    inds[inds == nbins] = nbins - 1

    p_i = np.array([np.sum(inds == i) for i in range(nbins)]) / len(angle)
    r_i = np.array([np.nanmean(activity[inds == i]) for i in range(nbins)])
    r_i = np.nan_to_num(r_i)
    r = np.nanmean(activity)

    with np.errstate(divide='ignore', invalid='ignore'):
        info_per_bin = p_i * (r_i / r) * np.log2((r_i / r) + 1e-12)
    info_per_bin[np.isnan(info_per_bin)] = 0

    return np.nansum(info_per_bin)

def permutation_test(activity, angle, n_perm=1000, nbins=10):
    info_real = compute_spatial_info(activity, angle, nbins=nbins)
    info_perm = np.zeros(n_perm)
    n = len(angle)
    for i in range(n_perm):
        shift = np.random.randint(n)
        activity_perm = np.roll(activity, shift)
        info_perm[i] = compute_spatial_info(activity_perm, angle, nbins=nbins)
    p_val = np.sum(info_perm >= info_real) / n_perm
    return info_real, info_perm, p_val

# ======== Analyse ========

angle_to_use = yaw_deg  # Choisir ici roll_deg ou pitch_deg si souhaité
nbins = 10
percentile_threshold = 90

results = []

for idx, neuron in enumerate(kept_units):
    activity = C_filtered[idx, :]
    valid_idx = ~np.isnan(angle_to_use)
    angle_vals = angle_to_use[valid_idx]
    activity_vals = activity[valid_idx]

    if len(activity_vals) < 10:
        continue

    # Test permutation
    info_real, info_perm, p_perm = permutation_test(activity_vals, angle_vals, n_perm=1000, nbins=nbins)
    threshold = np.percentile(info_perm, percentile_threshold)
    is_modulated_perm = info_real > threshold

    # Test KS (différence distribution activité / bins angle)
    bins = np.digitize(angle_vals, bins=np.linspace(np.nanmin(angle_vals), np.nanmax(angle_vals), nbins + 1)) - 1
    binned_means = np.array([np.nanmean(activity_vals[bins == i]) for i in range(nbins)])
    binned_means = np.nan_to_num(binned_means)
    ks_stat, ks_p = ks_2samp(activity_vals, binned_means)
    is_modulated_ks = ks_p < 0.05

    # Test Fisher (association activité forte vs bins)
    threshold_act = np.nanpercentile(activity_vals, 80)
    active = activity_vals > threshold_act
    spatial_bin = bins < (len(bins) // 2)
    contingency = pd.crosstab(active, spatial_bin)
    if contingency.shape == (2, 2):
        _, fisher_p = fisher_exact(contingency)
    else:
        fisher_p = np.nan
    is_modulated_fisher = fisher_p < 0.05 if not np.isnan(fisher_p) else False

    # Consensus (au moins 2 tests positifs)
    test_flags = [is_modulated_perm, is_modulated_ks, is_modulated_fisher]
    consensus = sum(test_flags) >= 2

    results.append({
        "Neuron": neuron,
        "Info_Spatial (Permutation)": info_real,
        f"Permutation_Threshold_{percentile_threshold}%": threshold,
        "P_Value_Permutation": p_perm,
        "Modulated_Permutation": is_modulated_perm,
        "KS_Stat": ks_stat,
        "P_Value_KS": ks_p,
        "Modulated_KS": is_modulated_ks,
        "P_Value_Fisher": fisher_p,
        "Modulated_Fisher": is_modulated_fisher,
        "Consensus_Modulated": consensus
    })

df_results = pd.DataFrame(results)
df_results.to_csv(dpath / "head_orientation_cells_results.csv", index=False)

print(f"✅ Analyse terminée, résultats sauvegardés dans {dpath / 'head_orientation_cells_results.csv'}")

display(
    df_results[df_results["Consensus_Modulated"] == True]
    .style
    .format(precision=4)
    .set_caption("Résultats détection head orientation cells - Consensus = True")
    .background_gradient(cmap="YlGnBu", subset=[
        "Info_Spatial (Permutation)", "P_Value_Permutation", "P_Value_KS", "P_Value_Fisher"
    ])
)


```
