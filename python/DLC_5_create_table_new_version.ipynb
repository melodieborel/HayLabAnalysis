```python
##### LOCAL #####

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import math
import os
from pathlib import Path
import re
from scipy.ndimage import median_filter
from scipy.interpolate import interp1d
from scipy.spatial import distance
import openpyxl

##############################################################################
# Choose the folder containing the .h5 files
##############################################################################

folder_path = "//10.69.168.1/crnldata/forgetting/Carla/Cheeseboard/" #will analyse all .avi files in this directory
counter=0
counterProbe=0
day=1
trial=1
previousmice=0
previous_session_time=0
Summary_table=pd.DataFrame()
Sholl_table=pd.DataFrame()

##############################################################################
# Define functions 
##############################################################################

def calculate_relative_distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

def calculate_distance_run(x_coords, y_coords):
    distances = np.sqrt(np.diff(x_coords) ** 2 + np.diff(y_coords) ** 2)
    for i in range(1, len(distances) - 1):
        if np.isnan(distances[i]):
            neighbors = [distances[i-1], distances[i+1]]
            distances[i] = np.mean([x for x in neighbors if not np.isnan(x)])
    total_distance_cm = np.nansum(distances) / pixel_to_cm  # Convert to cm
    return total_distance_cm, distances

def find_long_non_nan_sequences(arr, min_length=100):
    mask = ~np.isnan(arr)  # True for non-NaN values
    diff = np.diff(np.concatenate(([0], mask.astype(int), [0])))  # Add padding to detect edges
    starts = np.where(diff == 1)[0]  # Where a sequence starts
    ends = np.where(diff == -1)[0]   # Where a sequence ends
    sequences = [arr[start:end] for start, end in zip(starts, ends) if (end - start) > min_length]
    return sequences

def remove_outliers_median_filter(data, window=1):
    data = np.array(data, dtype=float)  # Ensure NumPy array with float type
    filtered_data = np.copy(data)  # Copy to avoid modifying original data
    half_window = window // 2
    for i in range(len(data)):
        # Define window range, ensuring it doesn't exceed bounds
        start = max(0, i - half_window)
        end = min(len(data), i + half_window + 1)
        # Extract local values in window
        local_values = data[start:end]
        # Check if the window contains at least one non-NaN value
        if np.all(np.isnan(local_values)):
            median_value = np.nan  # Keep NaN if no valid numbers
        else:
            median_value = np.nanmedian(local_values)  # Compute median ignoring NaNs
        # Replace only if the current value is not NaN
        if not np.isnan(data[i]):
            filtered_data[i] = median_value
    return filtered_data

def replace_high_speed_points_with_nan(x, y, speed_threshold):
    x = np.array(x, dtype='float')
    y = np.array(y, dtype='float')
    # Compute speed between consecutive points
    dx = np.diff(x)
    dy = np.diff(y)
    speeds = np.sqrt(dx**2 + dy**2)
    # Create mask for speed exceeding threshold
    high_speed_mask = speeds > speed_threshold
    # We mark i+1 as NaN if speed between them is too high
    x_out = x.copy()
    y_out = y.copy()
    for i in range(len(high_speed_mask)):
        if high_speed_mask[i]:
            # Only mark the faster of the two points
            if i > 0 and i < len(x) - 1:
                if speeds[i] > speeds[i - 1]:
                    x_out[i + 1] = np.nan
                    y_out[i + 1] = np.nan
                else:
                    x_out[i] = np.nan
                    y_out[i] = np.nan
    return x_out, y_out

def interpolate_2d_path(x, y, kind='linear', fill='extrapolate'):
    x = np.array(x, dtype='float')
    y = np.array(y, dtype='float')
    indices = np.arange(len(x))
    valid_mask = ~np.isnan(x) & ~np.isnan(y)
    if np.sum(valid_mask) < 2:
        raise ValueError("Not enough valid points to interpolate/extrapolate.")
    interp_x = interp1d(indices[valid_mask], x[valid_mask], kind=kind, fill_value=fill, bounds_error=False)
    interp_y = interp1d(indices[valid_mask], y[valid_mask], kind=kind, fill_value=fill, bounds_error=False)
    x_filled = x.copy()
    y_filled = y.copy()
    nan_mask = np.isnan(x) | np.isnan(y)
    x_filled[nan_mask] = interp_x(indices[nan_mask])
    y_filled[nan_mask] = interp_y(indices[nan_mask])
    return x_filled, y_filled

def limit_speed(x, y, max_speed):
    dx = np.diff(x.copy())
    dy = np.diff(y.copy())
    speeds = np.sqrt(dx**2 + dy**2)
    for i,t in enumerate(speeds):
        if t > max_speed:        
            x[i+1] = x[i] 
            y[i+1] = y[i] 
            if i+2 < len(x):   # ← décalé vers la droite !
                x[i+2] = x[i] 
                y[i+2] = y[i]
    return x, y

def remove_short_sequences(arr, max_len=10):
    arr = np.array(arr, dtype='float')
    result = arr.copy()
    is_value = ~np.isnan(arr)
    i = 0
    while i < len(arr):
        if is_value[i]:
            start = i
            while i < len(arr) and is_value[i]:
                i += 1
            end = i
            seq_len = end - start
            # Check if surrounded by NaNs and short enough
            if seq_len <= max_len:
                left_nan = (start == 0) or np.isnan(arr[start - 1])
                right_nan = (end == len(arr)) or np.isnan(arr[end])  # safe for edge
                if left_nan and right_nan:
                    result[start:end] = np.nan
        else:
            i += 1
    return result

def permutation_test(x, y, target_location, radius=8, table_radius=60, table_center=(0, 0), n_permutations=1000, seed=42):    
    # ========== INITIALIZATION ==========
    if seed is not None:
        np.random.seed(seed)
    x = np.asarray(x)
    y = np.asarray(y)
    target_location = np.asarray(target_location)
    center_x, center_y = table_center
    # ========== OBSERVED METRICS ==========
    # Calculate distance to target and identify frames inside zone
    dist_to_target = np.sqrt((x - target_location[0])**2 + (y - target_location[1])**2)
    inside_target = dist_to_target <= radius
    # Metric 1: First entry latency (seconds)
    observed_first_entry_frame = np.argmax(inside_target) if np.any(inside_target) else len(x)
    observed_first_entry_time = np.round(observed_first_entry_frame / frame_rate, 2)
    # Metric 2: Dwell time (% of trial)
    duration_trial = len(x) / frame_rate  # seconds
    observed_dwell = np.round((np.sum(inside_target) / frame_rate / duration_trial * 100), 2)
    # Metric 3: Entry rate (entries per 10 meters)
    total_distance, _ = calculate_distance_run(x, y)
    n_entries = np.sum(np.diff(inside_target.astype(int)) == 1)
    observed_entries_per_m = np.round(n_entries / total_distance * 10, 4)
    # ========== PERMUTATION TEST ==========
    random_dwell_times = []
    random_first_entry_times = []
    random_entries_per_m = []
    surrogate_targets = []
    while len(random_dwell_times) < n_permutations:
        # Generate random location uniformly within circular table
        r = table_radius * np.sqrt(np.random.rand())
        theta = 2 * np.pi * np.random.rand()
        rx = center_x + r * np.cos(theta)
        ry = center_y + r * np.sin(theta)
        # Skip if surrogate overlaps with reward zone (optional exclusion)
        if 'reward_x' in globals() and reward_x is not None:
            if calculate_relative_distance(rx, ry, reward_x, reward_y) <= reward_zone:
                continue
        # Calculate metrics for this surrogate target
        surrogate_targets.append([rx, ry])
        dist = np.sqrt((x - rx)**2 + (y - ry)**2)
        inside_zone = dist <= radius
        # Surrogate metric 1: First entry time
        first_entry_frame = np.argmax(inside_zone) if np.any(inside_zone) else len(x)
        first_entry_time = np.round(first_entry_frame / frame_rate, 2)
        # Surrogate metric 2: Dwell time
        dwell = np.round(np.sum(inside_zone) / frame_rate / duration_trial * 100, 2)
        # Surrogate metric 3: Entry rate
        n_entries_surr = np.sum(np.diff(inside_zone.astype(int)) == 1)
        entries_per_m = np.round(n_entries_surr / total_distance * 10, 4)
        random_dwell_times.append(dwell)
        random_first_entry_times.append(first_entry_time)
        random_entries_per_m.append(entries_per_m)
    
    # Convert to arrays
    surrogate_targets = np.array(surrogate_targets)
    random_dwell_times = np.array(random_dwell_times)
    random_first_entry_times = np.array(random_first_entry_times)
    random_entries_per_m = np.array(random_entries_per_m)
    
    # ========== CALCULATE P-VALUES ==========
    p_value_dwell = (np.sum(random_dwell_times >= observed_dwell) + 1) / (n_permutations + 1)
    p_value_entries = (np.sum(random_entries_per_m >= observed_entries_per_m) + 1) / (n_permutations + 1)
    p_value_first_entry = (np.sum(random_first_entry_times <= observed_first_entry_time) + 1) / (n_permutations + 1)
    
    # ========== COMBINED PERFORMANCE 3 METRIC WITH P-VALUE ==========
    # Calculate mean and std from random distribution (once)
    mean_dwell = np.mean(random_dwell_times)
    std_dwell = np.std(random_dwell_times)
    mean_entries = np.mean(random_entries_per_m)
    std_entries = np.std(random_entries_per_m)
    mean_latency = np.mean(random_first_entry_times)
    std_latency = np.std(random_first_entry_times)

    # Calculate z-scores for observed data (relative to random distribution)
    z_dwell_obs = (observed_dwell - mean_dwell) / std_dwell if std_dwell > 0 else 0
    z_entries_obs = (observed_entries_per_m - mean_entries) / std_entries if std_entries > 0 else 0
    z_latency_obs = -(observed_first_entry_time - mean_latency) / std_latency if std_latency > 0 else 0

    observed_perf_3m = z_dwell_obs + z_entries_obs + z_latency_obs
    observed_perf_2m = z_dwell_obs + z_entries_obs  
    # Calculate z-scores for each permutation (using SAME mean/std)
    z_dwell_perm = (random_dwell_times - mean_dwell) / std_dwell if std_dwell > 0 else np.zeros_like(random_dwell_times)
    z_entries_perm = (random_entries_per_m - mean_entries) / std_entries if std_entries > 0 else np.zeros_like(random_entries_per_m)
    z_latency_perm = -(random_first_entry_times - mean_latency) / std_latency if std_latency > 0 else np.zeros_like(random_first_entry_times)

    random_perf_3m = z_dwell_perm + z_entries_perm + z_latency_perm
    random_perf_2m = z_dwell_perm + z_entries_perm
    # Calculate p-value
    p_value_perf_3m = (np.sum(random_perf_3m >= observed_perf_3m) + 1) / (n_permutations + 1)
    p_value_perf_2m = (np.sum(random_perf_2m >= observed_perf_2m) + 1) / (n_permutations + 1)
    # ========== RETURN RESULTS ==========
    return {
        # Observed values
        'observed_dwell': observed_dwell,
        'observed_entries_per_m': observed_entries_per_m,
        'observed_first_entry_time': observed_first_entry_time,
        'observed_performance_3m': observed_perf_3m,
        'observed_performance_2m': observed_perf_2m,
        # Permutation distributions
        'random_dwell_times': random_dwell_times,
        'random_entries_per_m': random_entries_per_m,
        'random_first_entry_times': random_first_entry_times,
        'random_performance_3m': random_perf_3m,
        'random_performance_2m': random_perf_2m,
        # P-values
        'p_value_dwell': p_value_dwell,
        'p_value_entries': p_value_entries,
        'p_value_first_entry': p_value_first_entry,
        'p_value_performance_3m': p_value_perf_3m,
        'p_value_performance_2m': p_value_perf_2m,
    }

def apply_circular_mask(x, y, x_mask, y_mask, mask_radius):
    path_coords = np.column_stack((x, y))
    mask_center = np.array([x_mask, y_mask])
    dist_to_mask_center = distance.cdist(path_coords, [mask_center]).flatten()
    mask = dist_to_mask_center <= mask_radius
    return x[mask], y[mask]

def sholl_analysis(x, y, xr, yr, x_mask, y_mask, mask_radius=None, max_radius=100, step=10, buffer=20):
    if mask_radius is not None:
        x, y = apply_circular_mask(x, y, x_mask, y_mask, mask_radius)

    path_coords = np.column_stack((x, y))
    target = np.array([xr, yr])
    dist_to_target = distance.cdist(path_coords, [target]).flatten()

    radii = np.arange(step, max_radius + step, step)
    intersections = np.zeros(len(radii), dtype=int)

    for i, radius in enumerate(radii):
        crossings = np.where((dist_to_target[:-1] < radius) & (dist_to_target[1:] >= radius))[0]
        intersections[i] = len(crossings)
    
    return intersections, radii


##############################################################################
# Loop through all .h5 files in the folder and process them
##############################################################################

for root, dirs, files in os.walk(folder_path):
    h5_files = [f for f in files if f.endswith(('.h5'))]
    
    for h5_file in h5_files:
        filename = os.path.join(root, h5_file)
        filename = os.path.normpath(filename)

        ########################
        # Extract metadata from the file path
        ########################
        task_type = os.path.basename(Path(filename).parents[5])
        session_type = os.path.basename(Path(filename).parents[1])
        mice = os.path.basename(Path(filename).parents[3])
        sesstion_time = os.path.basename(Path(filename).parents[0])
        genotype = os.path.basename(Path(filename).parents[4])
        age= os.path.basename(Path(filename).parents[2])

        if "Cheeseboard" in task_type:
            print(f'Analyzing mouse {mice}, {age}, {session_type} session ')
            
            
            if mice == previousmice:   
                if session_type == previous_session_type : 
                    if sesstion_time==previous_session_time:
                        trial+=1
                    else:
                        day+=1
                        trial=1
                else: 
                    day=1
                    trial=1
            else:
                day=1
                trial=1

            ########################
            # Define parameters
            ########################
            pixel_to_cm = 5.05  
            reward_zone = 8 * pixel_to_cm if "Training" in filename or "Post" in filename else 20 * pixel_to_cm
            table_center_x, table_center_y = 920, 637
            table_radius = 606 / 2
            min_stay_at_reward_s = 2

            ########################
            # Load HDF5 file
            ########################        
            df = pd.read_hdf(filename)
            directory = os.path.dirname(filename)
            timestamps_path = Path(directory,'timeStamps.csv')
            if timestamps_path.exists():
                timestamps = pd.read_csv(timestamps_path)
                frame_rate = round(1/(np.mean(np.diff(timestamps.iloc[:,1]))/1000))  # fps
            else:
                frame_rate = 16  # fps /!\ CHANGE ACCORDING TO YOUR DATA
            min_stay_at_reward= min_stay_at_reward_s * frame_rate
            
            ########################
            # Load reward location
            ########################
            with open("//10.69.168.1/crnldata/forgetting/Carla/Cheeseboard/Reward_position.txt", "r") as file:
                text = file.read()  
            numbers = re.findall(r"[-+]?\d*\.\d+|\d+", text)
            reward_x, reward_y = map(float, numbers)

            ########################
            # Clean the signal
            ########################
            X0 = df.iloc[:, 0]
            Y0 = df.iloc[:, 1]

            # Remove uncertain location predictions (likelihood < 0.9)
            df.iloc[:, 0] = df.apply(lambda row: row.iloc[0] if row.iloc[2] > 0.5 else np.nan, axis=1)
            df.iloc[:, 1] = df.apply(lambda row: row.iloc[1] if row.iloc[2] > 0.5 else np.nan, axis=1)

            X = df.iloc[:, 0]
            Y = df.iloc[:, 1]

            # Separate the individual's positions into x and y coordinates
            individual_xO= np.array(X.values)
            individual_yO = np.array(Y.values)

                        # Define when the mouse is on the cheeseboard (start)
            for i, x in enumerate(individual_xO):
                y = individual_yO[i]
                if calculate_relative_distance(x, y, table_center_x, table_center_y) >= table_radius:
                    individual_xO[i] = np.nan
                    individual_yO[i] = np.nan

            individual_xOO = remove_short_sequences(individual_xO, max_len=3)
            individual_yOO = remove_short_sequences(individual_yO, max_len=3)

            if len(find_long_non_nan_sequences(individual_xOO)) == 0 or len(find_long_non_nan_sequences(individual_yOO)) == 0:
                print(f" Pas assez de données valides pour {mice}, session {session_type}, trial {trial} → ignoré")
                counter += 1
                previousmice = mice
                previous_session_time = sesstion_time
                previous_session_type = session_type
                continue

            x_start = find_long_non_nan_sequences(individual_xOO)[0][0] # first value of the first long non nan sequence
            y_start = find_long_non_nan_sequences(individual_yOO)[0][0] # first value of the first long non nan sequence

            start_frame = np.where(individual_xOO == x_start)[0][0].item()
            individual_xOO[:start_frame]=np.nan # remove any path before the real start
            individual_yOO[:start_frame]=np.nan # remove any path before the real start

            individual_x1, individual_y1 = replace_high_speed_points_with_nan(individual_xOO, individual_yOO, speed_threshold=10)

            for i in range(len(individual_x1)-1, 0, -1): # Find the last non-NaN value which is not isolated
                if not np.isnan(individual_x1[i]) and not np.isnan(individual_x1[i-1]):
                    last_frame = i
                    break

            individual_x2, individual_y2 = interpolate_2d_path(individual_x1[start_frame:last_frame], individual_y1[start_frame:last_frame], kind='nearest')
            individual_x3, individual_y3 = limit_speed(individual_x2, individual_y2, max_speed=20)

            individual_x = np.concatenate((individual_x1[:start_frame], individual_x3))
            individual_y = np.concatenate((individual_y1[:start_frame], individual_y3))

            ########################
            # Compute metrics
            ########################
            
            if timestamps_path.exists():
                start_time = timestamps.iloc[start_frame,1].item() / 1000
                end_time = timestamps.iloc[-1,1].item() / 1000
                duration_trial = end_time - start_time
            else:
                start_time = start_frame / frame_rate  
                end_time = last_frame / frame_rate   
                duration_trial = (last_frame - start_frame) / frame_rate

            total_distance, distances = calculate_distance_run(individual_x[start_frame:last_frame], individual_y[start_frame:last_frame])

            # Remove NaN values 
            individual_x_filt = individual_x[~np.isnan(individual_x)]        
            individual_y_filt = individual_y[~np.isnan(individual_y)]

            # Define when the mouse eats the reward (end)
            found_reward_frame_on_filtered = np.nan
            found_reward_time = np.nan
            enter_reward_zone = 0
            consecutive_count = 0
            timespent = 0
            reward_found = False
            for i, (x, y) in enumerate(zip(individual_x_filt, individual_y_filt)):
                if calculate_relative_distance(x, y, reward_x, reward_y) <= reward_zone:
                    timespent +=1
                    if consecutive_count == 0:
                        enter_reward_zone += 1
                        consecutive_count = 1
                else:
                    consecutive_count = 0   
                    timespent = 0
                if timespent > min_stay_at_reward:
                    found_reward_frame_on_filtered = (i - min_stay_at_reward)
                    reward_found =True
                    break

            if reward_found == False and len(X)/frame_rate<250:            
                found_reward_frame_on_filtered = np.nan
                found_reward_time = np.nan
                enter_reward_zone = 0
                consecutive_count = 0
                All_timespent=pd.DataFrame(columns=['timespent_in_zone', 'frame_i'])
                timespent = 0
                reward_found = True
                for i, (x, y) in enumerate(zip(individual_x_filt, individual_y_filt)):
                    if calculate_relative_distance(x, y, reward_x, reward_y) <= reward_zone:
                        timespent +=1
                        if consecutive_count == 0:
                            enter_reward_zone += 1
                            consecutive_count = 1
                        All_timespent.loc[enter_reward_zone, 'timespent_in_zone']=timespent
                        All_timespent.loc[enter_reward_zone, 'frame_i']=i-timespent
                    else:
                        consecutive_count = 0   
                        timespent = 0

                if All_timespent.empty:
                    print(f" {mice}, session {session_type}, trial {trial} → jamais entré dans la zone de récompense")
                    found_reward_frame_on_filtered = np.nan
                else:
                    found_reward_frame_on_filtered = int(All_timespent.loc[All_timespent['timespent_in_zone'].idxmax(), 'frame_i'])

            if reward_found and 'Training' in filename and not np.isnan(found_reward_frame_on_filtered):
                found_reward_frame1 = np.where(individual_x == individual_x_filt[int(found_reward_frame_on_filtered)])
                found_reward_frame2 = np.where(individual_y == individual_y_filt[int(found_reward_frame_on_filtered)])
                common = np.intersect1d(found_reward_frame1, found_reward_frame2)
                found_reward_frame = int(common[0])
                if timestamps_path.exists():
                    found_reward_time = timestamps.iloc[found_reward_frame,1].item() / 1000
                    latency = (found_reward_time - start_time)
                else :
                    found_reward_time=None
                    latency = (found_reward_frame - start_frame) / frame_rate
                found_reward_frame_int = int(found_reward_frame)
                distance_to_reward, distances_to_reward = calculate_distance_run(individual_x[start_frame:found_reward_frame_int], individual_y[start_frame:found_reward_frame_int])
            else:
                distance_to_reward=np.nan
                distances_to_reward=np.nan
                latency=np.nan

            crossings_per_m=round(enter_reward_zone / (round(total_distance) / 100), 2)

            # Define the time spent inside the reward zone
            time_spent_in_zone = 0
            for i, (x, y) in enumerate(zip(individual_x, individual_y)):
                if calculate_relative_distance(x, y, reward_x, reward_y) <= reward_zone:
                    time_spent_in_zone += 1
            time_spent_in_zone = time_spent_in_zone / frame_rate

            ########################
            # Create summary table
            ########################

            Summary_table.loc[counter, 'mice'] = mice
            Summary_table.loc[counter, 'age'] = age
            Summary_table.loc[counter, 'genotype'] = genotype
            Summary_table.loc[counter, 'session_type'] = session_type
            Summary_table.loc[counter, 'session'] = int(day)
            Summary_table.loc[counter, 'session_time'] = sesstion_time
            Summary_table.loc[counter, 'trial'] = int(trial)
            
            Summary_table.loc[counter, 'start_time'] = start_time
            Summary_table.loc[counter, 'end_time'] = end_time
            Summary_table.loc[counter, 'duration_trial_s'] = round(duration_trial,2)
            Summary_table.loc[counter, 'reward_zone_radius_cm'] = reward_zone/pixel_to_cm
            Summary_table.loc[counter, 'reward_location_pix'] = str([reward_x,reward_y])
            Summary_table.loc[counter, 'total_distance_cm'] = round(total_distance,2)
            Summary_table.loc[counter, 'average_speed_cm_s'] = round(total_distance/duration_trial,2)
            Summary_table.loc[counter, 'time_spent_in_reward_zone_s'] = round(time_spent_in_zone,2)
            Summary_table.loc[counter, 'time_spent_in_reward_zone_percent'] = round(time_spent_in_zone / duration_trial * 100,2)
            Summary_table.loc[counter, 'crossings'] = enter_reward_zone
            Summary_table.loc[counter, 'crossings_per_m'] = round(crossings_per_m, 2)
                
            if 'Training' in session_type:
                if found_reward_time is not None and not np.isnan(found_reward_time):
                    Summary_table.loc[counter, 'found_reward_time'] = found_reward_time
                    Summary_table.loc[counter, 'average_speed_to_reward_cm_s'] = round(np.nanmean(distances_to_reward)/pixel_to_cm*frame_rate,2) if not np.isnan(distances_to_reward).all() else np.nan
                    Summary_table.loc[counter, 'latency_to_reward_s'] = round(latency,2)
                    Summary_table.loc[counter, 'distance_to_reward_cm'] = round(distance_to_reward,2) if not np.isnan(distance_to_reward) else np.nan

            ########################
            # Permutation test for probe sessions
            ########################
            if 'Probe' in session_type or "Test" in session_type or "Habituation" in session_type:
                t=60*2
                results_2min = permutation_test(
                    individual_x[start_frame:int(start_frame+(t*frame_rate))], individual_y[start_frame:int(start_frame+(t*frame_rate))],
                    target_location=[reward_x, reward_y],
                    radius=reward_zone,
                    table_radius=table_radius,
                    table_center=(table_center_x, table_center_y),
                    n_permutations=1000,
                    seed=42)
                intersections_2min, radii_2min = sholl_analysis(individual_x[start_frame:int(start_frame+(t*frame_rate))], individual_y[start_frame:int(start_frame+(t*frame_rate))], reward_x, reward_y, table_center_x, table_center_y, mask_radius=table_radius, max_radius=table_radius*2, step=20)

                t=60*5
                results_5min = permutation_test(
                    individual_x[start_frame:int(start_frame+(t*frame_rate))], individual_y[start_frame:int(start_frame+(t*frame_rate))],
                    target_location=[reward_x, reward_y],
                    radius=reward_zone,
                    table_radius=table_radius,
                    table_center=(table_center_x, table_center_y),
                    n_permutations=1000,
                    seed=42)
                intersections_5min, radii_5min = sholl_analysis(individual_x[start_frame:int(start_frame+(t*frame_rate))], individual_y[start_frame:int(start_frame+(t*frame_rate))], reward_x, reward_y, table_center_x, table_center_y, mask_radius=table_radius, max_radius=table_radius*2, step=20)
                
                Summary_table.loc[counter, 'p_value_perf3m_2min'] = round(results_2min['p_value_performance_3m'], 4) if not np.isnan(results_2min['p_value_performance_3m']) else np.nan
                Summary_table.loc[counter, 'p_value_perf2m_2min'] = round(results_2min['p_value_performance_2m'], 4) if not np.isnan(results_2min['p_value_performance_2m']) else np.nan
                Summary_table.loc[counter, 'p_value_dwell_2min'] = round(results_2min['p_value_dwell'], 4) if not np.isnan(results_2min['p_value_dwell']) else np.nan
                Summary_table.loc[counter, 'p_value_entries_2min'] = round(results_2min['p_value_entries'], 4) if not np.isnan(results_2min['p_value_entries']) else np.nan
                Summary_table.loc[counter, 'p_value_first_entry'] = round(results_2min['p_value_first_entry'], 4) if not np.isnan(results_2min['p_value_first_entry']) else np.nan
                
                Summary_table.loc[counter, 'p_value_perf3m_3min'] = round(results_5min['p_value_performance_3m'], 4) if not np.isnan(results_5min['p_value_performance_3m']) else np.nan
                Summary_table.loc[counter, 'p_value_perf2m_3min'] = round(results_5min['p_value_performance_2m'], 4) if not np.isnan(results_5min['p_value_performance_2m']) else np.nan        
                Summary_table.loc[counter, 'p_value_dwell_3min'] = round(results_5min['p_value_dwell'], 4) if not np.isnan(results_5min['p_value_dwell']) else np.nan
                Summary_table.loc[counter, 'p_value_entries_3min'] = round(results_5min['p_value_entries'], 4) if not np.isnan(results_5min['p_value_entries']) else np.nan
                Summary_table.loc[counter, 'p_value_first_entry_3min'] = round(results_5min['p_value_first_entry'], 4) if not np.isnan(results_5min['p_value_first_entry']) else np.nan

                # Save Sholl data for this session
                Sholl_table.loc[counterProbe, 'mice'] = mice
                Sholl_table.loc[counterProbe, 'age'] = age
                Sholl_table.loc[counterProbe, 'genotype'] = genotype
                Sholl_table.loc[counterProbe, 'session_type'] = session_type
                Sholl_table.loc[counterProbe, 'session'] = int(day)
                Sholl_table.loc[counterProbe, 'session_time'] = sesstion_time
                Sholl_table.loc[counterProbe, 'trial'] = int(trial)
                
                for i in range(len(radii_2min)):
                    Sholl_table.loc[counterProbe, f'intersections_{int(radii_2min[i])}rad_2min'] = intersections_2min[i]
                for i in range(len(radii_5min)):
                    Sholl_table.loc[counterProbe, f'intersections_{int(radii_5min[i])}rad_3min'] = intersections_5min[i]
                counterProbe+=1

            previousmice = mice
            previous_session_time = sesstion_time
            previous_session_type = session_type
            counter+=1

filenameOut = f'{folder_path}/Summary_table_Cheeseboard.xlsx'
with pd.ExcelWriter(filenameOut) as writer:
    Summary_table.to_excel(writer, index=False)
filenameOut = f'{folder_path}/Sholl_analysis_table_Cheeseboard.xlsx'
with pd.ExcelWriter(filenameOut) as writer:
    Sholl_table.to_excel(writer, index=False)
```
