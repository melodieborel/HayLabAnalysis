```python
#Get data:
#Dict: keys = region coordinates; entries = pd dataframe with 3 columns (base, stim1, stim2) and nb_mice rows

histogram_allcx = {}
for coord in sorted(mice_coords[17], reverse=True):
    histogram_allcx[coord] = pd.DataFrame(np.nan, index=list(mice_coords.keys()), columns=["baseline", "stim_mid", "stim_vb"])

for mouse in mice_coords:
    for i, coord in enumerate(mice_coords[mouse]):
        baseline_sigma = []
        stim_mid_sigma = []
        stim_vb_sigma = []
        for freq in range(5,9):
            baseline_sigma.append(base_mid[mouse][i][freq,freq])
            baseline_sigma.append(base_vb[mouse][i][freq,freq])
            stim_mid_sigma.append(stim_mid[mouse][i][freq,freq])
            stim_vb_sigma.append(stim_vb[mouse][i][freq,freq])
        histogram_allcx[coord].loc[mouse] = [np.mean(baseline_sigma), np.mean(stim_mid_sigma), np.mean(stim_vb_sigma)]

#Delete HP entries
del histogram_allcx[(-1.7, 1.25)]
del histogram_allcx[(-1.7, -1.25)]

nb_mice = len(mice_coords)
means = []
sem = []
for key in histogram_allcx:
    region_means = np.mean(histogram_allcx[key], axis=0)
    region_sem = histogram_allcx[key].sem()
    means.append(region_means)
    sem.append(region_sem)
means = np.array(means)
sem = np.array(sem)
```

```python
# — example placeholders, replace with your real data —
regions   = list(histogram_allcx.keys())

plt.rcParams.update({
    "font.size":         12,
    "axes.titlesize":    16,
    "axes.labelsize":    14,
    "xtick.labelsize":   12,
    "xtick.major.width": 2,
    "ytick.labelsize":   12,
    "ytick.major.width": 2,
    "legend.fontsize":   12,
    "lines.linewidth":    2,
    "axes.linewidth":     2,
})

n_regions = len(regions)
bw = 0.2
x = np.arange(n_regions)
x_b   = x - bw
x_s1  = x
x_s2  = x + bw

fig, ax = plt.subplots(figsize=(12,6))

# 1) Bars with black edges
ax.bar(x_b,  means[:,0], width=bw, facecolor='lightgray',
       edgecolor='black', linewidth=2, label='Baseline')
ax.bar(x_s1, means[:,1], width=bw, facecolor='#a9c2d4',
       edgecolor='black', linewidth=2, label='Midline stim')
ax.bar(x_s2, means[:,2], width=bw, facecolor='#caeea1',
       edgecolor='black', linewidth=2, label='Som.sensory relay stim')

# 2) +SEM only (upwards) via vlines + horizontal caps
cap_width = bw * 0.6  # width of the little cap

for xpos, m, e in zip(x_b,  means[:,0], sem[:,0]):
    top = m + e
    ax.vlines(xpos, m, top, color='black', linewidth=2)
    ax.hlines(top, xpos - cap_width/2, xpos + cap_width/2,
              color='black', linewidth=2)

for xpos, m, e in zip(x_s1, means[:,1], sem[:,1]):
    top = m + e
    ax.vlines(xpos, m, top, color='black', linewidth=2)
    ax.hlines(top, xpos - cap_width/2, xpos + cap_width/2,
              color='black', linewidth=2)

for xpos, m, e in zip(x_s2, means[:,2], sem[:,2]):
    top = m + e
    ax.vlines(xpos, m, top, color='black', linewidth=2)
    ax.hlines(top, xpos - cap_width/2, xpos + cap_width/2,
              color='black', linewidth=2)

# 3) Lines for individual mouse results
for mouse_idx in range(nb_mice):
    for i, key in enumerate(histogram_allcx):
        yvals = histogram_allcx[key].iloc[mouse_idx][["baseline","stim_mid","stim_vb"]].values
        # plot the little polyline baseline→stim1→stim2 at the three x-positions:
        ax.plot(
            [x_b[i], x_s1[i], x_s2[i]],
            yvals,
            color='black',
            alpha=1,
            linewidth=2
        )

# 4) Final formatting
ax.set_xticks(x)
ax.set_xticklabels(regions, rotation=45, ha='right')
yt = ax.get_yticks()
ax.set_ylim(yt[0], yt[-1])
ax.set_ylabel("Spindle Power")
ax.set_title("Spindle Power by Region")
ax.legend(loc='upper right', ncol=3)

# Move x‐axis through y=0
ax.spines['bottom'].set_position(('data', 0))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# Stats test (according to chatgpt)
import numpy as np
from scipy.stats import wilcoxon
from statsmodels.stats.multitest import multipletests

# Sample data
baseline = micedata[:,0]
stimulated = micedata[:,2]

# Perform the Wilcoxon signed-rank test
stat, p_value = wilcoxon(baseline, stimulated)

# Bonferroni correction
# Since we are performing only one test, we need to adjust the alpha level
p_values_corrected = multipletests([p_value], alpha=0.05, method='bonferroni')

# Extract the corrected p-value
p_value_corrected = p_values_corrected[1][0]

# Print the results
print(f'Wilcoxon signed-rank test statistic: {stat}')
print(f'Original p-value: {p_value}')
print(f'Corrected p-value: {p_value_corrected}')
```

```python
large_bins_mid = np.zeros((7,8))
large_bins_vb = np.zeros((7,8))

for x, mouse in enumerate(mice):
    if mouse == 13 or 18:
        side = 0
    else:
        side = 1
    #Midline
    large_bins_mid[x, 0] = np.sum(globals()[f"mouse{mouse}_SWR"][side, 5:9, :5])
    large_bins_mid[x, 1] = np.sum(globals()[f"mouse{mouse}_SWR"][side, 5:9, 5:10])
    large_bins_mid[x, 2] = np.sum(globals()[f"mouse{mouse}_SWR"][side, 5:9, 10:15])

    afterstim = np.zeros((4, 5))
    afterstim_norm = np.zeros((4, 5))
    for a in range(5):
        afterstim[:, a] = np.sum((globals()[f"mouse{mouse}_SWR"][side, 5:9, 15+2*a], globals()[f"mouse{mouse}_SWR"][side, 5:9, 16+2*a]), axis=0)
    
    freqs = range(12, 20, 2)
    for b, freq in enumerate(freqs):
        afterstim_norm[b, :] = (afterstim[b, :]*10)/freq

    large_bins_mid[x, 3:] = np.sum(afterstim_norm, axis=0)

    #VB
    large_bins_vb[x, 0] = np.sum(globals()[f"mouse{mouse}_SWR"][side+2, 5:9, :5])
    large_bins_vb[x, 1] = np.sum(globals()[f"mouse{mouse}_SWR"][side+2, 5:9, 5:10])
    large_bins_vb[x, 2] = np.sum(globals()[f"mouse{mouse}_SWR"][side+2, 5:9, 10:15])

    afterstim = np.zeros((4, 5))
    afterstim_norm = np.zeros((4, 5))
    for a in range(5):
        afterstim[:, a] = np.sum((globals()[f"mouse{mouse}_SWR"][side+2, 5:9, 15+2*a], globals()[f"mouse{mouse}_SWR"][side+2, 5:9, 16+2*a]), axis=0)
    
    freqs = range(12, 20, 2)
    for b, freq in enumerate(freqs):
        afterstim_norm[b, :] = (afterstim[b, :]*10)/freq

    large_bins_vb[x, 3:] = np.sum(afterstim_norm, axis=0)

large_bins_mid[6, :] = np.mean(large_bins_mid[:6, :], axis=0)
large_bins_vb[6, :] = np.mean(large_bins_vb[:6, :], axis=0)

large_bins = np.concatenate((large_bins_mid, np.zeros((1,8)), large_bins_vb))
np.savetxt("Coord SWR large bins.csv", large_bins, fmt='%10.2f')
```
