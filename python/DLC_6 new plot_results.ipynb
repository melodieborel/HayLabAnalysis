# Plot mice performances on Cheeseboard Maze task

```python
##### LOCAL #####

import pandas as pd
from ipyfilechooser import FileChooser
import os
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib

%matplotlib widget

try:
    %store -r filename
    filename = os.path.dirname(filename)
except NameError:
    filename = "/crnldata/forgetting/Aurelie/CheeseboardExperiment/DAQ_data/AB/"

fc1 = FileChooser(filename, select_default=True, show_only_dirs=False, title="<b>Choose Summary Table excel file</b>")
display(fc1)

def update_file(chooser):
    global filename
    if chooser.selected:
        filename = chooser.selected
        %store filename

fc1.register_callback(update_file)
```

Results per day (n = mice)

```python
# -----------------------------
# Charger le fichier Excel
# -----------------------------
Summary_table = pd.read_excel(filename)

# -----------------------------
# Séparer Training et Test
# -----------------------------
Summary_table_Training = Summary_table[Summary_table['session_type'] == 'Training']
Nb_learning_sessions = np.size(Summary_table_Training['session'].unique())


# -----------------------------
# Créer figure et axes
# -----------------------------
fig, axs = plt.subplots(2, 3, figsize=(15, 6))


# -----------------------------
# Latency to reward
# -----------------------------
latency_to_reward_s = Summary_table_Training.pivot_table(
    index='mice', columns='session', values='latency_to_reward_s', aggfunc='mean'
)
mean_latency = np.nanmean(latency_to_reward_s.values, axis=0)
sem_latency = stats.sem(latency_to_reward_s.values, nan_policy='omit')

axs[0,0].plot(latency_to_reward_s.columns.values, latency_to_reward_s.values.T, marker='o', alpha=0.4, label=latency_to_reward_s.index)
axs[0,0].bar(latency_to_reward_s.columns.values, mean_latency, color="k", alpha=0.3)
axs[0,0].errorbar(latency_to_reward_s.columns.values, mean_latency, yerr=sem_latency, color="k", capsize=4)
axs[0,0].set(ylabel='Latency to reward (s)')
axs[0,0].set(xlabel='Learning days')
axs[0,0].legend()

# -----------------------------
# Distance to reward
# -----------------------------
distance_to_reward_cm = Summary_table_Training.pivot_table(
   index='mice', columns='session', values='distance_to_reward_cm', aggfunc='mean'
)
mean_dist = np.nanmean(distance_to_reward_cm.values, axis=0)
sem_dist = stats.sem(distance_to_reward_cm.values, nan_policy='omit')

axs[0,1].plot(distance_to_reward_cm.columns.values, distance_to_reward_cm.values.T, marker='o', alpha=0.4)
axs[0,1].bar(distance_to_reward_cm.columns.values, mean_dist, color="k", alpha=0.3)
axs[0,1].errorbar(distance_to_reward_cm.columns.values, mean_dist, yerr=sem_dist, color="k", capsize=4)
axs[0,1].set(ylabel='Distance to reward (cm)')
axs[0,1].set(xlabel='Learning days')

# -----------------------------
# Average speed (cm/s)
# -----------------------------
average_speed = Summary_table_Training.pivot_table(
        index='mice', columns='session', values='average_speed_cm_s', aggfunc='mean'
)

mean_speed = np.nanmean(average_speed.values, axis=0)
sem_speed =stats.sem(average_speed.values, nan_policy='omit')

axs[0,2].plot(average_speed.columns.values, average_speed.values.T, marker='o', alpha=0.4)
axs[0,2].bar(average_speed.columns.values, mean_speed, color="k", alpha=0.3)
axs[0,2].errorbar(average_speed.columns.values, mean_speed, yerr=sem_speed, color="k", capsize=4)
axs[0,2].set(ylabel='Average speed (cm.s-1)')
axs[0,2].set(xlabel='Learning days')


# -----------------------------
# Time spent at reward (s)
# -----------------------------
time_spent_at_reward_s = Summary_table_Training.pivot_table(
        index='mice',
    columns='session',
    values='time_spent_in_reward_zone_s',
    aggfunc='mean'
)

mean_s=np.nanmean(time_spent_at_reward_s.values, axis=0)
sem_s=stats.sem(time_spent_at_reward_s.values, nan_policy='omit')

axs[1,0].plot(time_spent_at_reward_s.columns.values,time_spent_at_reward_s.values.T, alpha=0.4, marker='o')
axs[1,0].bar(time_spent_at_reward_s.columns.values, mean_s, color="k", alpha=0.3)
axs[1,0].errorbar(time_spent_at_reward_s.columns.values, mean_s, yerr=sem_s, color="k", capsize=4)
axs[1,0].set(ylabel='Time spent at reward (s)')
axs[1,0].set(xlabel='Days')

# -----------------------------
# Time spent at reward (%)
# -----------------------------
perc_time_spent_at_reward = Summary_table_Training.pivot_table(
    index='mice',
    columns='session',
    values='time_spent_in_reward_zone_percent',
    aggfunc='mean'
)

mean_perc = np.nanmean(perc_time_spent_at_reward.values, axis=0)
sem_perc = stats.sem(perc_time_spent_at_reward.values, nan_policy='omit')

axs[1,1].plot(perc_time_spent_at_reward.columns.values, perc_time_spent_at_reward.values.T, alpha=0.4, marker='o')
axs[1,1].bar(perc_time_spent_at_reward.columns.values, mean_perc, color="k", alpha=0.3)
axs[1,1].errorbar(perc_time_spent_at_reward.columns.values, mean_perc, yerr=sem_perc, color="k", capsize=4)
axs[1,1].set(ylabel='Time spent at reward (%)')
axs[1,1].set(xlabel='Days')

# -----------------------------
# Crossings per m
# -----------------------------
crossings_per_m = Summary_table_Training.pivot_table(
    index='mice',
    columns='session',
    values='crossings_per_m',
    aggfunc='mean'
)
mean_cross=np.nanmean(crossings_per_m.values, axis=0)
sem_cross=stats.sem(crossings_per_m.values, nan_policy='omit')

axs[1,2].plot(crossings_per_m.columns.values, crossings_per_m.values.T, alpha=0.4, marker='o')
axs[1,2].bar(crossings_per_m.columns.values, mean_cross, color="k", alpha=0.3)
axs[1,2].errorbar(crossings_per_m.columns.values, mean_cross, yerr=sem_cross, color="k", capsize=4)

axs[1,2].set(ylabel='Crossings per m')
axs[1,2].set(xlabel='Days')

plt.tight_layout()
plt.show()
```

```python
# -----------------------------
# Charger le fichier Excel
# -----------------------------
Summary_table = pd.read_excel(filename)

# -----------------------------
# Séparer Training et Test
# -----------------------------
Summary_table_Habituation = Summary_table[Summary_table['session_type'] == 'Habituation']
Summary_table_Probe = Summary_table[Summary_table['session_type'] == 'Probe']
Nb_learning_sessions = np.size(Summary_table_Habituation['session'].unique())


# -----------------------------
# Créer figure et axes
# -----------------------------
fig, axs = plt.subplots(1, 3, figsize=(15, 6))

# Créer un DataFrame avec Hab et Probe pour chaque souris
#df_hab = pd.DataFrame({'mice': crossings_hab.index, 'mean_hab': mean_hab})
#df_probe = pd.DataFrame({'mice': crossings_probe.index, 'probe': crossings_probe.values})

# Garder seulement les souris présentes dans les deux conditions
#df_plot = pd.merge(df_hab, df_probe, on='mice', how='inner')

# -----------------------------
# Crossings per m
# -----------------------------
crossings_hab = Summary_table_Habituation.pivot_table(
    index='mice',
    columns='session',
    values='crossings_per_m',
    aggfunc='mean'
)
mean_cross_hab = np.nanmean(crossings_hab.values, axis=1)

crossings_probe = Summary_table_Probe.set_index('mice')['crossings_per_m']

df_plot = pd.merge(
    pd.DataFrame({'mice': crossings_hab.index, 'mean_cross_hab': mean_cross_hab}),
    pd.DataFrame({'mice': crossings_probe.index, 'probe': crossings_probe.values}),
    on='mice', how='inner'
)

data_to_plot = df_plot[['mean_cross_hab', 'probe']].values.T

mean1 = np.nanmean(data_to_plot, axis=1)
sem1 = [stats.sem(data_to_plot[0], nan_policy='omit'),
       stats.sem(data_to_plot[1], nan_policy='omit')]

axs[0].plot([1,2], data_to_plot, marker='o', alpha=0.4, label=crossings_per_m.index)
axs[0].bar([1,2], mean1, color=["k","r"], alpha=0.3)
axs[0].errorbar([1,2], mean1, yerr=sem1, color="k", capsize=4)
axs[0].set_xticks([1,2])
axs[0].set_xlabel('Sessions')
axs[0].set_xticklabels(['Habituation','Probe'])
axs[0].set_ylabel('Crossings per m')
axs[0].legend()


# -----------------------------
# Time in reward zone 
# -----------------------------
time_hab = Summary_table_Habituation.pivot_table(
    index='mice',
    columns='session',
    values='time_spent_in_reward_zone_s',
    aggfunc='mean'
)
mean_time_hab = np.nanmean(time_hab .values, axis=1)

time_probe = Summary_table_Probe.set_index('mice')['time_spent_in_reward_zone_s']

df_plot2 = pd.merge(
    pd.DataFrame({'mice': time_hab.index, 'mean_time_hab': mean_time_hab}),
    pd.DataFrame({'mice': time_probe.index, 'probe': time_probe.values}),
    on='mice', how='inner'
)

data_to_plot = df_plot2[['mean_time_hab', 'probe']].values.T

mean2 = np.nanmean(data_to_plot, axis=1)
sem2 = [stats.sem(data_to_plot[0], nan_policy='omit'),
       stats.sem(data_to_plot[1], nan_policy='omit')]

for i, mouse in enumerate(df_plot2['mice']):
    axs[1].plot([1,2], data_to_plot[:, i], marker='o', alpha=0.4, label=mouse)
axs[1].bar([1,2], mean2, color=["k","r"], alpha=0.3)
axs[1].errorbar([1,2], mean2, yerr=sem2, color="k", capsize=4)
axs[1].set_xticks([1,2])
axs[1].set_xlabel('Sessions')
axs[1].set_xticklabels(['Habituation','Probe'])
axs[1].set_ylabel('time_spent_in_reward_zone_s')


# -----------------------------
# Percent time in reward zone
# -----------------------------

perc_hab = Summary_table_Habituation.pivot_table(
    index='mice',
    columns='session',
    values='time_spent_in_reward_zone_percent',
    aggfunc='mean'
)
mean_perc_hab = np.nanmean(perc_hab .values, axis=1)

perc_probe = Summary_table_Probe.set_index('mice')['time_spent_in_reward_zone_percent']

df_plot3 = pd.merge(
    pd.DataFrame({'mice': perc_hab.index, 'mean_perc_hab': mean_perc_hab}),
    pd.DataFrame({'mice': perc_probe.index, 'probe': perc_probe.values}),
    on='mice', how='inner'
)

data_to_plot = df_plot3[['mean_perc_hab', 'probe']].values.T

mean3 = np.nanmean(data_to_plot, axis=1)
sem3 = [stats.sem(data_to_plot[0], nan_policy='omit'),
       stats.sem(data_to_plot[1], nan_policy='omit')]

for i, mouse in enumerate(df_plot3['mice']):
    axs[2].plot([1,2], data_to_plot[:, i], marker='o', alpha=0.4, label=mouse)
axs[2].bar([1,2], mean3, color=["k","r"], alpha=0.3)
axs[2].errorbar([1,2], mean3, yerr=sem3, color="k", capsize=4)
axs[2].set_xticks([1,2])
axs[2].set_xlabel('Sessions')
axs[2].set_xticklabels(['Habituation','Probe'])
axs[2].set_ylabel('time_spent_in_reward_zone_percent')


plt.show()
```
