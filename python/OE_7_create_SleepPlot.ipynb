```python
with open("//10.69.168.1/crnldata/forgetting/Carla/SleepRecordings/EMGHypno_dict - Copie.pkl", 'rb') as f:
    EMGHypno_dict = pickle.load(f)
Summary_table = pd.read_excel("//10.69.168.1/crnldata/forgetting/Carla/SleepRecordings/Summary_table - Copie.xlsx")


for mouse in EMGHypno_dict.keys():
    # --- PLOT ---
    fig, axs = plt.subplots(1,len(EMGHypno_dict[mouse].keys()),figsize=(15,2))
    axs = axs.flatten()
    
    genotype=Summary_table[Summary_table['ImplantColorCode']==mouse]['Genotype'].unique().tolist()[0]
    sex=Summary_table[Summary_table['ImplantColorCode']==mouse]['Sex'].unique().tolist()[0]

    for age in EMGHypno_dict[mouse].keys():

        EMGfilt_DS = zscore(EMGHypno_dict[mouse][age]['filtEMG'])
        starttime = EMGHypno_dict[mouse][age]['StartTime']
        Hypno= EMGHypno_dict[mouse][age]['Hypno']

        start = pd.to_datetime(starttime)
        fs = 100          # 100 Hz
        N = len(EMGfilt_DS)
        dt = 1.0 / fs     # sample spacing in seconds

        # generate timestamps
        offsets = np.arange(N) * dt
        datetimes = start + pd.to_timedelta(offsets, unit='s')

        y = EMGfilt_DS

        # --- BUILD TICKS for every day's 07:00 and 19:00 ---
        unique_days = sorted({d.date() for d in datetimes})
        ticks = []
        for d in unique_days:
            ticks.append(pd.to_datetime(f"{d} 07:00:00"))
            ticks.append(pd.to_datetime(f"{d} 19:00:00"))
        ticks = [t for t in ticks if datetimes.min() <= t <= datetimes.max()]

        ax = axs[age]

        ax.plot(datetimes, y, linewidth=0.5)
        ax.set_ylabel("EMG & Hypno")
        ax.set_xticks(sorted(ticks))
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
        ax.plot(datetimes, Hypno*10-50, linewidth=0.5)
        plt.xticks(rotation=45)
        ax.set_xlim(datetimes.min(), datetimes.max())
        ax.set_title(f'{age}', fontsize=10)


    plt.tight_layout()
    fig.suptitle(f'{genotype} {sex} {mouse}', fontsize=10)
    plt.tight_layout()
    plt.savefig(f'//10.69.168.1/crnldata/forgetting/Carla/SleepRecordings/{mouse}/EMGHypno_{mouse}.png', dpi=300, bbox_inches='tight', facecolor='white')

```
