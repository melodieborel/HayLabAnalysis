Define functions

```python
def calculate_relative_distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

def calculate_distance_run(x_coords, y_coords):
    distances = np.sqrt(np.diff(x_coords) ** 2 + np.diff(y_coords) ** 2)
    for i in range(1, len(distances) - 1):
        if np.isnan(distances[i]):
            neighbors = [distances[i-1], distances[i+1]]
            distances[i] = np.mean([x for x in neighbors if not np.isnan(x)])
    total_distance_cm = np.nansum(distances) / pixel_to_cm  # Convert to cm
    return total_distance_cm, distances

def find_long_non_nan_sequences(arr, min_length=100):
    mask = ~np.isnan(arr)  # True for non-NaN values
    diff = np.diff(np.concatenate(([0], mask.astype(int), [0])))  # Add padding to detect edges
    starts = np.where(diff == 1)[0]  # Where a sequence starts
    ends = np.where(diff == -1)[0]   # Where a sequence ends
    sequences = [arr[start:end] for start, end in zip(starts, ends) if (end - start) > min_length]
    return sequences

def remove_outliers_median_filter(data, window=1):
    data = np.array(data, dtype=float)  # Ensure NumPy array with float type
    filtered_data = np.copy(data)  # Copy to avoid modifying original data
    half_window = window // 2
    for i in range(len(data)):
        # Define window range, ensuring it doesn't exceed bounds
        start = max(0, i - half_window)
        end = min(len(data), i + half_window + 1)
        # Extract local values in window
        local_values = data[start:end]
        # Check if the window contains at least one non-NaN value
        if np.all(np.isnan(local_values)):
            median_value = np.nan  # Keep NaN if no valid numbers
        else:
            median_value = np.nanmedian(local_values)  # Compute median ignoring NaNs
        # Replace only if the current value is not NaN
        if not np.isnan(data[i]):
            filtered_data[i] = median_value
    return filtered_data

def replace_high_speed_points_with_nan(x, y, speed_threshold):
    x = np.array(x, dtype='float')
    y = np.array(y, dtype='float')
    # Compute speed between consecutive points
    dx = np.diff(x)
    dy = np.diff(y)
    speeds = np.sqrt(dx**2 + dy**2)
    # Create mask for speed exceeding threshold
    high_speed_mask = speeds > speed_threshold
    # We mark i+1 as NaN if speed between them is too high
    x_out = x.copy()
    y_out = y.copy()
    for i in range(len(high_speed_mask)):
        if high_speed_mask[i]:
            # Only mark the faster of the two points
            if i > 0 and i < len(x) - 1:
                if speeds[i] > speeds[i - 1]:
                    x_out[i + 1] = np.nan
                    y_out[i + 1] = np.nan
                else:
                    x_out[i] = np.nan
                    y_out[i] = np.nan
    return x_out, y_out

def interpolate_2d_path(x, y, kind='linear', fill='extrapolate'):
    x = np.array(x, dtype='float')
    y = np.array(y, dtype='float')
    indices = np.arange(len(x))
    valid_mask = ~np.isnan(x) & ~np.isnan(y)
    if np.sum(valid_mask) < 2:
        raise ValueError("Not enough valid points to interpolate/extrapolate.")
    interp_x = interp1d(indices[valid_mask], x[valid_mask], kind=kind, fill_value=fill, bounds_error=False)
    interp_y = interp1d(indices[valid_mask], y[valid_mask], kind=kind, fill_value=fill, bounds_error=False)
    x_filled = x.copy()
    y_filled = y.copy()
    nan_mask = np.isnan(x) | np.isnan(y)
    x_filled[nan_mask] = interp_x(indices[nan_mask])
    y_filled[nan_mask] = interp_y(indices[nan_mask])
    return x_filled, y_filled

def limit_speed(x, y, max_speed):
    dx = np.diff(x.copy())
    dy = np.diff(y.copy())
    speeds = np.sqrt(dx**2 + dy**2)
    for i,t in enumerate(speeds):
        if t > max_speed:        
            x[i+1] = x[i] 
            y[i+1] = y[i] 
            x[i+2] = x[i] 
            y[i+2] = y[i] 
    return x, y

def remove_short_sequences(arr, max_len=10):
    arr = np.array(arr, dtype='float')
    result = arr.copy()
    is_value = ~np.isnan(arr)
    i = 0
    while i < len(arr):
        if is_value[i]:
            start = i
            while i < len(arr) and is_value[i]:
                i += 1
            end = i
            seq_len = end - start
            # Check if surrounded by NaNs and short enough
            if seq_len <= max_len:
                left_nan = (start == 0) or np.isnan(arr[start - 1])
                right_nan = (end == len(arr)) or np.isnan(arr[end])  # safe for edge
                if left_nan and right_nan:
                    result[start:end] = np.nan
        else:
            i += 1
    return result
```

## Run tests on Probe days

Permutation test with 3 metrics (Time spent in zone, Entry rate, Latency to first entry)

```python
def permutation_test(
    x, y, target_location, radius=8, table_radius=60, table_center=(0, 0),
    n_permutations=1000, seed=42): 
    """
    Perform permutation test to assess spatial memory performance.
    
    Tests three metrics against random target locations:
    1. Time spent in zone (% of trial duration)
    2. Entry rate (entries per meter traveled)
    3. Latency to first entry (seconds)
    """
    
    # ========== INITIALIZATION ==========
    if seed is not None:
        np.random.seed(seed)

    x = np.asarray(x)
    y = np.asarray(y)
    target_location = np.asarray(target_location)
    center_x, center_y = table_center
    
    # ========== OBSERVED METRICS ==========
    # Calculate distance to target and identify frames inside zone
    dist_to_target = np.sqrt((x - target_location[0])**2 + (y - target_location[1])**2)
    inside_target = dist_to_target <= radius
    
    # Metric 1: First entry latency (seconds)
    observed_first_entry_frame = np.argmax(inside_target) if np.any(inside_target) else len(x)
    observed_first_entry_time = np.round(observed_first_entry_frame / frame_rate, 2)
    
    # Metric 2: Dwell time (% of trial)
    duration_trial = len(x) / frame_rate  # seconds
    observed_dwell = np.round((np.sum(inside_target) / frame_rate / duration_trial * 100), 2)
    
    # Metric 3: Entry rate (entries per 10 meters)
    total_distance, _ = calculate_distance_run(x, y)
    n_entries = np.sum(np.diff(inside_target.astype(int)) == 1)
    observed_entries_per_m = np.round(n_entries / total_distance * 10, 4)

    # ========== PERMUTATION TEST ==========
    random_dwell_times = []
    random_first_entry_times = []
    random_entries_per_m = []
    surrogate_targets = []
    
    while len(random_dwell_times) < n_permutations:
        # Generate random location uniformly within circular table
        r = table_radius * np.sqrt(np.random.rand())
        theta = 2 * np.pi * np.random.rand()
        rx = center_x + r * np.cos(theta)
        ry = center_y + r * np.sin(theta)
        
        # Skip if surrogate overlaps with reward zone (optional exclusion)
        if 'reward_x' in globals() and reward_x is not None:
            if calculate_relative_distance(rx, ry, reward_x, reward_y) <= reward_zone:
                continue
        
        # Calculate metrics for this surrogate target
        surrogate_targets.append([rx, ry])
        dist = np.sqrt((x - rx)**2 + (y - ry)**2)
        inside_zone = dist <= radius
        
        # Surrogate metric 1: First entry time
        first_entry_frame = np.argmax(inside_zone) if np.any(inside_zone) else len(x)
        first_entry_time = np.round(first_entry_frame / frame_rate, 2)
        
        # Surrogate metric 2: Dwell time
        dwell = np.round(np.sum(inside_zone) / frame_rate / duration_trial * 100, 2)
        
        # Surrogate metric 3: Entry rate
        n_entries_surr = np.sum(np.diff(inside_zone.astype(int)) == 1)
        entries_per_m = np.round(n_entries_surr / total_distance * 10, 4)
        
        random_dwell_times.append(dwell)
        random_first_entry_times.append(first_entry_time)
        random_entries_per_m.append(entries_per_m)
    
    # Convert to arrays
    surrogate_targets = np.array(surrogate_targets)
    random_dwell_times = np.array(random_dwell_times)
    random_first_entry_times = np.array(random_first_entry_times)
    random_entries_per_m = np.array(random_entries_per_m)
    
    # ========== CALCULATE P-VALUES ==========
    p_value_dwell = (np.sum(random_dwell_times >= observed_dwell) + 1) / (n_permutations + 1)
    p_value_entries = (np.sum(random_entries_per_m >= observed_entries_per_m) + 1) / (n_permutations + 1)
    p_value_first_entry = (np.sum(random_first_entry_times <= observed_first_entry_time) + 1) / (n_permutations + 1)
    
    
    # ========== COMBINED PERFORMANCE METRIC WITH P-VALUE ==========

    # Calculate mean and std from random distribution (once)
    mean_dwell = np.mean(random_dwell_times)
    std_dwell = np.std(random_dwell_times)
    mean_entries = np.mean(random_entries_per_m)
    std_entries = np.std(random_entries_per_m)
    mean_latency = np.mean(random_first_entry_times)
    std_latency = np.std(random_first_entry_times)

    # Calculate z-scores for observed data (relative to random distribution)
    z_dwell_obs = (observed_dwell - mean_dwell) / std_dwell if std_dwell > 0 else 0
    z_entries_obs = (observed_entries_per_m - mean_entries) / std_entries if std_entries > 0 else 0
    z_latency_obs = -(observed_first_entry_time - mean_latency) / std_latency if std_latency > 0 else 0

    observed_perf_3m = z_dwell_obs + z_entries_obs + z_latency_obs
    observed_perf_2m = z_dwell_obs + z_entries_obs 

    # Calculate z-scores for each permutation (using SAME mean/std)
    z_dwell_perm = (random_dwell_times - mean_dwell) / std_dwell if std_dwell > 0 else np.zeros_like(random_dwell_times)
    z_entries_perm = (random_entries_per_m - mean_entries) / std_entries if std_entries > 0 else np.zeros_like(random_entries_per_m)
    z_latency_perm = -(random_first_entry_times - mean_latency) / std_latency if std_latency > 0 else np.zeros_like(random_first_entry_times)

    random_perf_3m = z_dwell_perm + z_entries_perm + z_latency_perm
    random_perf_2m = z_dwell_perm + z_entries_perm

    # Calculate p-value
    p_value_perf_3m = (np.sum(random_perf_3m >= observed_perf_3m) + 1) / (n_permutations + 1)
    p_value_perf_2m = (np.sum(random_perf_2m >= observed_perf_2m) + 1) / (n_permutations + 1)

    print(f"Combined performance p-value (3 metrics): {p_value_perf_3m:.4f}")
    print(f"Combined performance p-value (2 metrics): {p_value_perf_2m:.4f}")

    
    # ========== VISUALIZATION ==========
    fig, axes = plt.subplots(1, 4, figsize=(14, 3))
    
    # Plot 1: Dwell time distribution
    axes[0].hist(random_dwell_times, bins=30, alpha=0.7, color='steelblue', label='Random targets')
    axes[0].axvline(observed_dwell, color='red', linestyle='--', linewidth=2, label=f'Observed: {observed_dwell}%')
    axes[0].set_title(f'Time in Zone\np = {p_value_dwell:.4f}', fontsize=11, fontweight='bold')
    axes[0].set_xlabel('Time in zone (%)', fontsize=10)
    axes[0].set_ylabel('Frequency', fontsize=10)
    axes[0].legend(fontsize=9)
    
    # Plot 2: Entry rate distribution
    axes[1].hist(random_entries_per_m, bins=30, alpha=0.7, color='seagreen', label='Random targets')
    axes[1].axvline(observed_entries_per_m, color='red', linestyle='--', linewidth=2, 
                    label=f'Observed: {observed_entries_per_m}')
    axes[1].set_title(f'Entry Rate\np = {p_value_entries:.4f}', fontsize=11, fontweight='bold')
    axes[1].set_xlabel('Entries per m', fontsize=10)
    axes[1].set_ylabel('Frequency', fontsize=10)
    axes[1].legend(fontsize=9)
    
    # Plot 3: First entry latency distribution
    axes[2].hist(random_first_entry_times, bins=30, alpha=0.7, color='mediumpurple', label='Random targets')
    axes[2].axvline(observed_first_entry_time, color='red', linestyle='--', linewidth=2, 
                    label=f'Observed: {observed_first_entry_time}s')
    axes[2].set_title(f'First Entry Latency\np = {p_value_first_entry:.4f}', fontsize=11, fontweight='bold')
    axes[2].set_xlabel('Time to first entry (s)', fontsize=10)
    axes[2].set_ylabel('Frequency', fontsize=10)
    axes[2].legend(fontsize=9)
    
    # Plot 4: Trajectory and targets
    axes[3].plot(x, y, '-', color='gray', alpha=0.6, linewidth=0.8, label='Path')
    axes[3].scatter(x[inside_target], y[inside_target], color='red', s=5, 
                    alpha=0.8, label='Inside target zone', zorder=3)
    axes[3].scatter(surrogate_targets[:, 0], surrogate_targets[:, 1], 
                    s=3, color='lightblue', alpha=0.4, label=f'{n_permutations} random targets')
    
    # Draw table boundary and target zone
    table_circle = plt.Circle(table_center, table_radius, fill=False, 
                              color='black', linewidth=2, linestyle='-')
    target_circle = plt.Circle(target_location, radius, fill=False, 
                               color='blue', linewidth=2, linestyle='--')
    axes[3].add_artist(table_circle)
    axes[3].add_artist(target_circle)
    axes[3].scatter(*target_location, color='blue', s=80, marker='*', 
                    zorder=5, edgecolors='black', linewidths=0.5, label='Target')
    
    axes[3].set_aspect('equal')
    axes[3].invert_yaxis()
    axes[3].set_title('Spatial Distribution', fontsize=11, fontweight='bold')
    axes[3].set_xlabel('X position (cm)', fontsize=10)
    axes[3].set_ylabel('Y position (cm)', fontsize=10)
    axes[3].legend(fontsize=8, loc='upper right')
    
    plt.tight_layout()
    plt.show()
    
    # ========== RETURN RESULTS ==========
    return {
        # Observed values
        'observed_dwell': observed_dwell,
        'observed_entries_per_m': observed_entries_per_m,
        'observed_first_entry_time': observed_first_entry_time,
        'observed_performance_3m': observed_perf_3m,
        'observed_performance_2m': observed_perf_2m,
        # Permutation distributions
        'random_dwell_times': random_dwell_times,
        'random_entries_per_m': random_entries_per_m,
        'random_first_entry_times': random_first_entry_times,
        'random_performance_3m': random_perf_3m,
        'random_performance_2m': random_perf_2m,
        # P-values
        'p_value_dwell': p_value_dwell,
        'p_value_entries': p_value_entries,
        'p_value_first_entry': p_value_first_entry,
        'p_value_performance_3m': p_value_perf_3m,
        'p_value_performance_2m': p_value_perf_2m,
    }


print("#################################################")
t=60*2
print(f"On the first {t/60} minutes of the trial:")
results= permutation_test(
    individual_x[start_frame:int(start_frame+(t*frame_rate))], individual_y[start_frame:int(start_frame+(t*frame_rate))],
    target_location=[reward_x, reward_y],
    radius=reward_zone,
    table_radius=table_radius,
    table_center=(table_center_x, table_center_y),
    n_permutations=1000,
    seed=42
)

print("#################################################")
t=60*3
print(f"On the first {t/60} minutes of the trial:")
results_full = permutation_test(
    individual_x[start_frame:int(start_frame+(t*frame_rate))], individual_y[start_frame:int(start_frame+(t*frame_rate))],
    target_location=[reward_x, reward_y],
    radius=reward_zone,
    table_radius=table_radius,
    table_center=(table_center_x, table_center_y),
    n_permutations=1000,
    seed=42
)
print("#################################################")
```

Sholl analysis

```python
def apply_circular_mask(x, y, x_mask, y_mask, mask_radius):
    path_coords = np.column_stack((x, y))
    mask_center = np.array([x_mask, y_mask])
    dist_to_mask_center = distance.cdist(path_coords, [mask_center]).flatten()
    mask = dist_to_mask_center <= mask_radius
    return x[mask], y[mask]

def sholl_analysis(x, y, xr, yr, x_mask, y_mask, mask_radius=None, max_radius=100, step=10, buffer=20):
    if mask_radius is not None:
        x, y = apply_circular_mask(x, y, x_mask, y_mask, mask_radius)

    path_coords = np.column_stack((x, y))
    target = np.array([xr, yr])
    dist_to_target = distance.cdist(path_coords, [target]).flatten()

    radii = np.arange(step, max_radius + step, step)
    intersections = np.zeros(len(radii), dtype=int)

    for i, radius in enumerate(radii):
        crossings = np.where((dist_to_target[:-1] < radius) & (dist_to_target[1:] >= radius))[0]
        intersections[i] = len(crossings)

    # Plot path and circles

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.5))

    ax1.plot(x, y, 'b-', alpha=1, label="Path")
    ax1.scatter(xr, yr, c='red', s=50, label="Target")

    if mask_radius is not None:
        mask_circle = mpath.Path.circle((x_mask, y_mask), radius=mask_radius)
        mask_patch = mpatches.PathPatch(mask_circle, transform=ax1.transData, clip_on=True, facecolor='none', edgecolor='none')

    for radius in radii:
        circle = mpath.Path.circle((xr, yr), radius=radius)
        circle_patch = mpatches.PathPatch(circle, transform=ax1.transData, facecolor='none', edgecolor='r', linestyle='--', alpha=0.8, clip_on=True)
        if mask_radius is not None:
            circle_patch.set_clip_path(mask_patch)
        ax1.add_patch(circle_patch)

    if mask_radius is not None:
        mask_boundary = plt.Circle((x_mask, y_mask), mask_radius, color='black', fill=False, linestyle='-', linewidth=1)
        ax1.add_patch(mask_boundary)
        ax1.set_xlim(x_mask - mask_radius - buffer, x_mask + mask_radius + buffer)
        ax1.set_ylim(y_mask - mask_radius - buffer, y_mask + mask_radius + buffer)

    ax1.set_aspect('equal')
    ax1.invert_yaxis()
    ax1.set_xlabel("X Coordinate")
    ax1.set_ylabel("Y Coordinate")
    ax1.legend()

    # Plot Sholl analysis
    ax2.plot(radii, intersections, 'o-', color='blue')
    ax2.set_xlabel("Distance from Reward (cm)")
    ax2.set_ylabel("Number of Intersections")

    plt.tight_layout()
    plt.show()

    return intersections, radii

t=60*2
print(f"On the first {t/60} minutes of the trial:")
intersections_2min, radii_2min = sholl_analysis(individual_x[start_frame:int(start_frame+(t*frame_rate))], individual_y[start_frame:int(start_frame+(t*frame_rate))], reward_x, reward_y, table_center_x, table_center_y, mask_radius=table_radius, max_radius=table_radius*2, step=20)

t=60*5
print(f"On the first {t/60} minutes of the trial:")
intersections_5min, radii_5min = sholl_analysis(individual_x[start_frame:int(start_frame+(t*frame_rate))], individual_y[start_frame:int(start_frame+(t*frame_rate))], reward_x, reward_y, table_center_x, table_center_y, mask_radius=table_radius, max_radius=table_radius*2, step=20)
```
