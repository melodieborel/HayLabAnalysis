```python
import pandas as pd

A=pd.read_excel('//10.69.168.1/crnldata/waking/audrey_hay/L1imaging/AnalysedMarch2023/Gaelle/AB_GlobalAnalysis/AVG_VigSt_2024-10-03_17_36_46_AB_newScor_CGP/Si_L23_Baseline_CGP.xlsx')
```

```python
def replace_values(arr):
    result = []
    for value in arr:
        if value < -0.5:
            result.append(0) #REM
        elif -0.5 <= value <= 0.5:
            result.append(1) #Non sel
        elif 0.5 < value: 
            result.append(2) # NREM
        else:
            result.append(value)  # Keep value if it's outside the defined ranges
    return result
```

```python
# Apply the function
A['Baseline_N'] = replace_values(A['Baseline'])
A['CGP_N'] = replace_values(A['CGP'])

# Count the number of occurrences of each combination of col1 and col2
combination_counts = A.groupby(['Baseline_N', 'CGP_N']).size().reset_index(name='count')
```
